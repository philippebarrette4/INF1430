```r
library(dplyr)
library(readr)
```

```r
rawDF <- read_csv("../../data/dataverse_files/Electricity_B2E.csv", show_col_types = FALSE)
glimpse(rawDF)
```

```r
B2E <- rawDF %>% 
    mutate(
        timeStamp   = as.POSIXct(unix_ts, origin = "1970-01-01 12:00:00"),
        date        = as.Date(format(as.POSIXct(unix_ts, origin = "1970-01-01 12:00:00"), format = "%Y-%m-%d"), format = "%Y-%m-%d"),
        year        = format(as.POSIXct(unix_ts, origin = "1970-01-01 12:00:00"), format = "%Y"),
        month       = format(as.POSIXct(unix_ts, origin = "1970-01-01 12:00:00"), format = "%m"),
        day         = format(as.POSIXct(unix_ts, origin = "1970-01-01 12:00:00"), format = "%d"),
        hour        = format(as.POSIXct(unix_ts, origin = "1970-01-01 12:00:00"), format = "%H"),
        minute      = format(as.POSIXct(unix_ts, origin = "1970-01-01 12:00:00"), format = "%M")
    ) %>%
    select(-unix_ts)
```

```r
B2E_2 <- B2E %>%
    arrange(timeStamp) %>%
    mutate(
        Pt_kW = (Pt - lag(Pt)) * DPF / 1000, 
        Qt_kW = (Qt - lag(Qt)) * DPF / 1000,
        St_kW = (St - lag(St)) * DPF / 1000 
    )
    
B2E_2 %>%
    group_by(year) %>% 
    summarise(
        P_Total = sum(P) / 1000,
        Q_Total = sum(Q) / 1000,
        S_Total = sum(S) / 1000,
        Pt_Total = sum(Pt_kW, na.rm = TRUE),
        Qt_Total = sum(Qt_kW, na.rm = TRUE),
        St_Total = sum(St_kW, na.rm = TRUE),
        St_test = (max(St) - min(St)) / 1000,
        Qt_test = (max(Qt) - min(Qt)) / 1000,
        Pt_test = (max(Pt) - min(Pt)) / 1000
    )
```
