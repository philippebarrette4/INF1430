```r
library(dplyr)
library(readr)
library(lubridate)
```

```r
raw1 <- read_csv("../../data/Electricity_B1E.csv", show_col_types = FALSE)
raw2 <- read_csv("../../data/Electricity_B2E.csv", show_col_types = FALSE)
raw3 <- read_csv("../../data/Electricity_BME.csv", show_col_types = FALSE)
```

```r
B1E <- raw1 %>% 
    mutate(
        timeStamp   = as.POSIXct(unix_ts, origin = "1970-01-01 12:00:00"),
        date        = as.Date(format(as.POSIXct(unix_ts, origin = "1970-01-01 12:00:00"), format = "%Y-%m-%d"), format = "%Y-%m-%d"),
        year        = format(as.POSIXct(unix_ts, origin = "1970-01-01 12:00:00"), format = "%Y"),
        month       = format(as.POSIXct(unix_ts, origin = "1970-01-01 12:00:00"), format = "%m"),
        day         = format(as.POSIXct(unix_ts, origin = "1970-01-01 12:00:00"), format = "%d"),
        hour        = format(as.POSIXct(unix_ts, origin = "1970-01-01 12:00:00"), format = "%H"),
        minute      = format(as.POSIXct(unix_ts, origin = "1970-01-01 12:00:00"), format = "%M")
    ) %>%
    select(-unix_ts)

B2E <- raw2 %>% 
    mutate(
        timeStamp   = as.POSIXct(unix_ts, origin = "1970-01-01 12:00:00"),
        date        = as.Date(format(as.POSIXct(unix_ts, origin = "1970-01-01 12:00:00"), format = "%Y-%m-%d"), format = "%Y-%m-%d"),
        year        = format(as.POSIXct(unix_ts, origin = "1970-01-01 12:00:00"), format = "%Y"),
        month       = format(as.POSIXct(unix_ts, origin = "1970-01-01 12:00:00"), format = "%m"),
        day         = format(as.POSIXct(unix_ts, origin = "1970-01-01 12:00:00"), format = "%d"),
        hour        = format(as.POSIXct(unix_ts, origin = "1970-01-01 12:00:00"), format = "%H"),
        minute      = format(as.POSIXct(unix_ts, origin = "1970-01-01 12:00:00"), format = "%M")
    ) %>%
    select(-unix_ts)

BME <- raw3 %>% 
    mutate(
        timeStamp   = as.POSIXct(unix_ts, origin = "1970-01-01 12:00:00"),
        date        = as.Date(format(as.POSIXct(unix_ts, origin = "1970-01-01 12:00:00"), format = "%Y-%m-%d"), format = "%Y-%m-%d"),
        year        = format(as.POSIXct(unix_ts, origin = "1970-01-01 12:00:00"), format = "%Y"),
        month       = format(as.POSIXct(unix_ts, origin = "1970-01-01 12:00:00"), format = "%m"),
        day         = format(as.POSIXct(unix_ts, origin = "1970-01-01 12:00:00"), format = "%d"),
        hour        = format(as.POSIXct(unix_ts, origin = "1970-01-01 12:00:00"), format = "%H"),
        minute      = format(as.POSIXct(unix_ts, origin = "1970-01-01 12:00:00"), format = "%M")
    ) %>%
    select(-unix_ts)
```

```r
B1E %>% 
    mutate(
        `RP (kW)` = (S * DPF) / 1000
    ) %>% 
    group_by(year) %>%
    summarise(
        kW_Total = sum(`RP (kW)`),
        St_Total = (max(St) - min(St)) / 1000
    ) %>%
    ungroup() %>%
    summarise(avgkWH = mean(St_Total))
```

```r
B2E %>%
    group_by(year) %>%
    summarise(
        kW_Total = (max(St) - min(St)) / 1000
    )
```

```r
BME %>%
    group_by(year) %>%
    summarise(
        kW_Total = (max(St) - min(St)) / 1000
    )
```
